```r
support_df <- function(tre1, tre2, label){
    sp1 <- tre1$node.label[-1][-1]
    sp2 <- tre2$node.label[-1][-1]
    sp1 <- round(as.numeric(gsub(paste0(".*", label, "=([0-9.]*).*"), "\\1", sp1)), 4)
    sp2 <- round(as.numeric(gsub(paste0(".*", label, "=([0-9.]*).*"), "\\1", sp2)), 4)
    l1 <- length(sp1)
    l2 <- length(sp2)
    sp1 <- data.frame(Tree = rep("One class", l2), Support = sp1)
    sp2 <- data.frame(Tree = rep("Mixture", l1), Support = sp2)
    sp <- rbind(sp1, sp2)
    sp$Support <- as.numeric(sp$Support)
    return(sp)
}
sp_untrimmed <- support_df(Inferred_tree_single_untrimmed, Inferred_tree_mix_untrimmed, "pp1")
sp_gappyout <- support_df(Inferred_tree_single_gappyout, Inferred_tree_mix_gappyout, "pp1")
sp_gt <- support_df(Inferred_tree_single_gt, Inferred_tree_mix_gt, "pp1")
sp_untrimmed$Type <- "untrimmed"
sp_gappyout$Type <- "gappyout"
sp_gt$Type <- "gt"
sp <- rbind(sp_untrimmed, sp_gappyout, sp_gt)
# Create a new column for the interval
sp$Interval <- cut(sp$Support, breaks = seq(0, 1, by = 0.1), include.lowest = TRUE)
sp$Tree <- factor(sp$Tree, levels = c("One class", "Mixture"))
sp$Type <- factor(sp$Type, levels = c("untrimmed", "gt", "gappyout"))

# Convert the intervals to percentages
levels(sp$Interval) <- paste0(as.character(seq(0, 90, by = 10)), "~", as.character(seq(10, 100, by = 10)), "%")

# Plot the histogram
plot_pp <- ggplot(sp, aes(x = Interval, fill = Tree)) +
  geom_bar(position = "dodge") +
  theme_Publication() +
  labs(fill = "Model", x = "Posterior probability interval", y = "Count")+
  facet_wrap(~Type, ncol = 1)

sp2_untrimmed <- support_df(Inferred_tree_single_untrimmed, Inferred_tree_mix_untrimmed, "q1")
sp2_gappyout <- support_df(Inferred_tree_single_gappyout, Inferred_tree_mix_gappyout, "q1")
sp2_gt <- support_df(Inferred_tree_single_gt, Inferred_tree_mix_gt, "q1")
sp2_untrimmed$Type <- "untrimmed"
sp2_gappyout$Type <- "gappyout"
sp2_gt$Type <- "gt"
sp2 <- rbind(sp2_untrimmed, sp2_gappyout, sp2_gt)
# Create a new column for the interval
sp2$Interval <- cut(sp2$Support, breaks = seq(0, 1, by = 0.1), include.lowest = TRUE)
sp2$Tree <- factor(sp2$Tree, levels = c("One class", "Mixture"))
sp2$Type <- factor(sp2$Type, levels = c("untrimmed", "gt", "gappyout"))

# Convert the intervals to percentages
levels(sp2$Interval) <- paste0(as.character(seq(0, 90, by = 10)), "~", as.character(seq(10, 100, by = 10)), "%")

plot_qcf <- ggplot(sp2, aes(x = Interval, fill = Tree)) +
  geom_bar(position = "dodge") +
  theme_Publication() +
  labs(fill = "Model", x = "qCF sore interval", y = "Count")+
  facet_wrap(~Type, ncol = 1)

options(repr.plot.width = 14, repr.plot.height = 9)
plot_pp + plot_qcf +
  plot_annotation(tag_levels = "A") & 
  theme(plot.tag = element_text(size = 15))
```

```r
library(TreeDist)
trim_trees <- function(tree1, tree2) {
  # Get the labels of the tips of both trees
  taxa_tree1 <- tree1$tip.label
  taxa_tree2 <- tree2$tip.label

  # Find the intersection of the labels
  taxa_intersect <- intersect(taxa_tree1, taxa_tree2)

  # Trim both trees to only include the intersecting labels
  drop_taxa_tree1 <- setdiff(taxa_tree1, taxa_intersect)
  drop_taxa_tree2 <- setdiff(taxa_tree2, taxa_intersect)
  
  tree1_trimmed <- drop.tip(tree1, drop_taxa_tree1)
  tree2_trimmed <- drop.tip(tree2, drop_taxa_tree2)

  # Return the trimmed trees
  return(list(tree1 = tree1_trimmed, tree2 = tree2_trimmed))
}

# Define the function
tree_distance <- function(tree_list, tree_names) {
  # Check the input arguments
  if (length(tree_list) != 2) {
    stop("tree_list must have exactly two trees")
  }
  if (length(tree_names) != 2) {
    stop("tree_names must have exactly two names")
  }
  
  # Get the trees and their names
  tree1 <- tree_list[[1]]
  tree2 <- tree_list[[2]]
  tree1_name <- tree_names[1]
  tree2_name <- tree_names[2]
  
  # Check if the trees have the same labels
  if (!setequal(tree1$tip.label, tree2$tip.label)) {
    stop("the trees must have the same labels")
  }
  if (length(tree1$tip.label) == 0) {
    stop("the trees must have at least one tip")
  }
  
  # Calculate distances
  nRF_dist <-  round(TreeDistance(tree1, tree2),4)
  SPR_dist <- round(SPR.dist(tree1, tree2), 4)
  path_dist <- round(path.dist(tree1, tree2, check.labels = TRUE, use.weight = TRUE), 4)
  
  # Create a data frame with the results
  result_df <- data.frame(
    N_taxa = length(tree1$tip.label),
    tree_1 = tree1_name,
    tree_2 = tree2_name,
    nRF_distance = nRF_dist,
    SPR_distance = SPR_dist
  )
  
  # Return the data frame
  return(result_df)
}


# Define the function
measure_tree_distances <- function(tree_left, tree_right, tree_name_left, tree_name_right) {
  # Initialize an empty data frame to store the results
  result_df <- data.frame(
    N_taxa = numeric(0),
    tree_1 = character(0),
    tree_2 = character(0),
    nRF_distance = numeric(0),
    SPR_distance = numeric(0)
  )
  
  # Loop over all pairs of trees
  for (i in seq_along(tree_left)) {
    for (j in seq_along(tree_right)) {
      # Trim the trees to their common labels
      trimmed_trees <- trim_trees(tree_left[[i]], tree_right[[j]])
      
      # Measure the distances between the trimmed trees
      distances <- tree_distance(trimmed_trees, c(tree_name_left[i], tree_name_right[j]))
      
      # Add the distances to the result data frame
      result_df <- rbind(result_df, distances)
    }
  }
  
  # Return the result data frame
  return(result_df)
}
```

```r
trans_treefile <- function(phy, label, path){
    node_info <- phy$node.label
    label_values <- round(as.numeric(gsub(paste0(".*", label, "=([0-9.]*).*"), "\\1", node_info)), 4)
    phy$node.label <- label_values
    phy$edge.length[is.na(phy$edge.length)] <- 1
    write.nexus(phy, file = path)
}

for (i in 1:length(my_trees)){
    trans_treefile(my_trees[[i]],"q1" , paste(my_tree_names[i], "_transfered.nw", sep = ""))
}
```
